# Scrape all of Byredo

```python
import time

results = []

for i, url in enumerate(urls):
    print(f"Scraping {i+1}/{len(urls)}: {url}")
    
    data = scraper.scrape_perfume_details(url)
    
    if data:
        results.append(data)
    
    # random delay between 3-7 seconds
    delay = random.uniform(3, 7)
    print(f"Waiting {delay:.1f} seconds...")
    time.sleep(delay)

print(f"\nDone! Scraped {len(results)} perfumes")
df = pd.DataFrame(results)
df.head()
```

```python
df_byredo = df[df['brand_name'] == 'Byredo']
```

```python
df_byredo.to_csv('byredo_data.csv', index=False)
```
